### Question 1a


**What is the most common start station?**

```r
start_stations <- chic %>%   
  group_by(Start.Station) %>%
  summarise(n = n()) %>%
  arrange(desc(n)) %>%
  filter(n > 100) # graph had too much x variables wihout filter
ggplot(start_stations, aes(Start.Station, n))+
    geom_col()+
    labs(title = 'Chicago')+
    coord_flip()
```

```r
start_stations <- nyc %>%   
  group_by(Start.Station) %>%
  summarise(n = n()) %>%
  arrange(desc(n)) %>%
  filter(n > 300) # graph had too much x variables wihout filter

ggplot(start_stations, aes(Start.Station, n))+
  geom_col()+
  labs(title = 'New York City')+
    coord_flip()
```

```r
start_stations <- wash %>%   
  group_by(Start.Station) %>%
  summarise(n = n()) %>%
  arrange(desc(n)) %>%
  filter(n > 1000) # graph had too much x variables wihout filter

ggplot(start_stations, aes(Start.Station, n))+
  geom_col()+
  labs(title = 'Washington DC')+
  coord_flip()
```

The most common start station for Chicago is Streeter Dr & Grand Ave
  The most common start station for New York City is Pershing Square North
  The most common start station for Washington is Columbus Circle/ Union Station
  
  The filter had to be used because the x variables were too many and unreadable. Coord_flip() was used to display the station names in their completeness

# Question 1c

**What is the most frequent combination of start and end stations?**

```r
routes <- chic %>%
  group_by(Start.Station,End.Station)%>%
  summarise(route_count = n()) %>%
  arrange(desc(route_count)) %>%
  filter(route_count>10)

ggplot(routes,aes(Start.Station, End.Station))+
  geom_point(aes(size = route_count))+
  scale_size(range = c(1,10))+
  labs(title = 'Chicago')+
  theme(axis.text.x = element_text(angle = 90))
```

```r
routes <- nyc %>%
  group_by(Start.Station,End.Station)%>%
  summarise(route_count = n()) %>%
  arrange(desc(route_count)) %>%
  filter(route_count>15)

ggplot(routes,aes(Start.Station, End.Station))+
  geom_point(aes(size = route_count))+
  labs(title = 'New York City')+
  theme(axis.text.x = element_text(angle = 90))
```

```r
routes <- wash %>%
  group_by(Start.Station,End.Station)%>%
  summarise(route_count = n()) %>%
  arrange(desc(route_count)) %>%
  filter(route_count>100)

ggplot(routes,aes(Start.Station, End.Station))+
  geom_point(aes(size = route_count))+
  labs(title = 'Washington DC')+
  theme(axis.text.x = element_text(angle = 90))
```

In Chicago, the most frequent route was from Streeter Dr/ Grand Ave and back to same location
In New York City, the most frequent route was from E7 St & Avenue A to Cooper Square & E7 St. Notably is that closely following is round trip rides from Central Park/ 6TH Ave and back.
In Washington, the most common route is the route from Jefferson Dr & 14th St SW and back. Closely following is another round trip from Smithsonian-National Mall / Jefferson Dr & 12th St SW and back

**What is the total travel time for users in different cities**

```r
tot_travel_chicago <- chic%>%
  summarise(total_travel_chicago  = sum(Trip.Duration)) #summation of trip durations



tot_travel_nyc <- nyc%>%
  summarise(total_travel_nyc  = sum(Trip.Duration))



tot_travel_wash <- wash%>%
  summarise(total_travel_wash  = sum(Trip.Duration))

cbind_travel_time <- cbind(tot_travel_wash,tot_travel_nyc,tot_travel_chicago)

              # Create new table with info on all cities from cbind_travel_time
total_travel_time <- c(109883545 ,49490073,8084781)
city <- c( 'Washington', 'New_York_City', 'Chicago')

              # Combine vectors into data frame
df1 <- data.frame(city, total_travel_time)

head(df1)

ggplot(df1)+
  geom_col(aes(city, total_travel_time))+
  labs(title = 'Comparison of Total Travel Times')
```

**What are the counts of each gender?**

```r

gender_types_chic <- chic %>%
  group_by(Gender) %>%
  summarise(count = n())

gender_types_nyc <- nyc %>%
  group_by(Gender) %>%
  summarise(count = n())

ggplot(gender_types_chic)+
  geom_col(aes(Gender, count))+
  labs(title = 'Users By Gender: Chicago')


ggplot(gender_types_nyc)+
  geom_col(aes(Gender, count))+
  labs(title = 'Users By Gender: New York City')
```

## Question 3c

**What are the earliest, most recent, most common year of birth (only available for NYC and Chicago)?**

```r
frequent_year_chicago <- chic%>%
  group_by(Birth.Year)%>%
  summarise(count = n())%>%
  arrange(desc(count))%>%
  filter(count>300)     #to display the topmost frequent years of birthh only

head(frequent_year_chicago)

summary(chic$Birth.Year)            #this gives the min and max years, no mode (most frequent year)
```

The earliest year in Chicago is 1899. The latest is 2002. The most frequent is 1989

```r
frequent_year_nyc <- nyc%>%
  group_by(Birth.Year)%>%
  summarise(count = n())%>%
  arrange(desc(count))%>%
  filter(count>300)     #to display the topmost frequent years of birthh only

head(frequent_year_nyc)

summary(nyc$Birth.Year)            #this gives the min and max years, no mode (most frequent year)
```

The earliest year in NYC is 1885. The latest is 2001. The most frequent is 1989, same as in Chicago
